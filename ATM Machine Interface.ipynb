```python
#This program is improvisied from the 'secret password'program and 'amount withdrawal deposit' program(from session 4) 
#This program is kind of ATM machine program where user can perform certain operation with respect to
#his/her bank account with the help of their userID and password.
#This program is meant for user front only.
#This program is meant for 5 customers only with their bank details.
import time
import numpy as np
dict2={'@grawal':'A01X9','roy@123':'RB25S','rav_321':'RU68R','anu&78':'AN34C','farh*345':'FN26D'} #In dict2,keys-->userid,values-->passwords
dict1={'A01X9':'Likhit','RB25S':'Bhushan','RU68R':'Ranveer','AN34C':'Anuj','FN26D':'Farahan'}       #In dict1, keys-->passwords,values--> customer name
dict3={'@grawal':5000,'roy@123':8000,'rav_321':4000,'anu&78':7000,'farh*345':6500}   #userid with corresponding saving account balance
dict4={'@grawal':1000,'roy@123':2000,'rav_321':5000,'anu&78':3000,'farh*345':7200}      # userid with corresponding current account balance
lst=[]
lst1=[]
n=0
while True:
    if n==3:
        break
    userid=input('Enter your userid: ')       #refer dict2 for userid
    if userid not in dict2:                   #giving chance user to enter userid if userid is not in dict2 
        print('No Record Found')
        print('Do you want to try again? YES OR NO ')  
        ans=input().lower()
        if ans=='yes':                      # if yes, then prompt user input again and again unless enter correct userid  
            continue
        else:
            break                             # Program will end if entered anything other than 'YES' in ans prompt
    while n<3:                                # giving 3 chances to user to enter correct password
        password=input('Enter your secret password: ')
        print('Loading....')
        time.sleep(1)
        if password in dict1:
            print('\n')                              # comparing key which is user password from dict1 with password user inserted
            print('Welcome to the system',dict1[password])
            time.sleep(2)
        else:
            print('Incorrect password')       # after 3 iteration, loop will end
            n+=1 
            continue                          # n will jump on previous loop,therefore, break n at n=3
            
#only 2 type of account has been used in this problem: savings and current 
#enter 1 for Savings account and 2 for current account
# This code is for withdraw,deposit amount and to check transcation history and remaining balance from balance inquiry  
        while True:                            
            print('Select your Account type: ')
            print('1. Savings Account                                  2. Current Account')                                 
            acc=int(input('Enter 1 or 2 '))
            
            # after selecting savings account
            if acc==1:           
                print('1.Withdraw                2.Deposit             3.Balance inquiry')
                choice= int(input('Enter the number 1,2 or 3: '))
                
                # For withdrawing amount from savings account
                while choice==1: 
                    ans=int(input('Enter the withdrawal amount: '))# initial balance of user is stored in dict3
                    time.sleep(2)
                    if dict3[userid]<ans:                         # loop will continue to run unless withdrawal           
                        print('Insufficient Balance')             #amount is less than balance available in savings account
                        continue    
                    else:
                        dict3[userid]-=ans                        # storing withdawal amount in lst for showing transaction history
                        lst.append(-ans)                          
                    print('Do you want receipt of the transaction? yes/no')
                    rec=input().lower()                           
                    if rec=='yes':                                 
                        print('Transcation Details:')             # showing transaction details to user
                        print('Withdrawal amount',ans)
                        print('Remaining balance',dict3[userid])
                        break
                    elif rec=='no':
                        break
                # For depositing amount to savings account        
                while choice==2:
                    ans=int(input('Enter the deposit amount '))    
                    dict3[userid]+=ans                     # incrementing the balance of the user after amount deposited     
                    lst.append(+ans)                       # storing the amount in the lst for transaction history
                    print('Do you want receipt of the transaction? yes/no')
                    rec1=input().lower()
                    time.sleep(2)
                    if rec1=='yes':
                        print('Transcation Details:')
                        print('Deposited amount',ans)
                        print('Total balance:',dict3[userid])
                        break
                    elif rec1=='no':
                        break  
                        
                # For showing transaction history and total balance of savings account        
                if choice==3:
                    for i in lst:
                        if i<0:
                            print('Amount withdrawn : ',-1*i) # This will shows withdrawal amount history of savings account
                        else:
                            print('Amount deposited : ',i)     # this will shows history of deposited amount of savings account
                    print('Total balance: ',dict3[userid])
                print('Do you want to do another transaction?') # user can start the transaction again from selecting account type
                choice=input('yes or no: ').lower()
                if choice=='yes':
                    continue
                elif choice =='no':
                    print('Thank you for banking with us')
                    print('Rate our service from 1 to 5 :')
                    rate=int(input())
                    if rate<3:                            #giving option to user for suggestions if user rating is less than 3 out of 5
                        print('sorry for inconvinence :-( , would you like to add any suggestion')
                        sugg=input('Add suggestion... ')
                        print('Thank you')
                    elif rate>3:
                        print('Thank you :-D')
                    break
            
            # after selecting current account
            if acc==2:
                print('1.Withdraw                2.Deposit             3.Balance inquiry')
                choice= int(input('Enter the number 1,2 or 3 '))
                
                # For withdrawing amount from current account
                while choice==1:
                    ans1=int(input('Enter the withdrawal amount ')) # initial balance of user is stored in dict4
                    time.sleep(2)
                    if dict4[userid]<ans1:                         # loop will continue to run unless withdrawal 
                        print('Insufficient Balance')              #amount is less than balance available in current
                        continue    
                    else:
                        dict4[userid]-=ans1                        # storing withdawal amount in lst for showing transaction history
                        lst1.append(-ans1)
                    print('Do you want reciept of the transaction? yes/no')
                    rec=input().lower()
                    if rec=='yes':
                        print('Transcation Details:')                 # showing transaction details to user
                        print('Withdrawal amount : ',ans1)
                        print('Remaining balance : ',dict4[userid])
                        break
                    elif rec=='no':
                        break
                        
                # For depositing amount to current account        
                while choice==2:
                    ans1=int(input('Enter the deposit amount '))
                    time.sleep(2)
                    dict4[userid]+=ans1                      # incrementing the balance of the user after amount deposited
                    lst1.append(+ans1)                        # storing the amount in the lst for current account transaction history
                    print('Do you want reciept of the transaction? yes/no')
                    rec=input().lower()
                    if rec=='yes':
                        print('Transcation Details:')
                        print('Amount Deposited : ',ans1)
                        print('Total balance :',dict4[userid])
                        break
                    elif rec=='no':
                        break
                        
                # For showing transaction history and total balance of current account         
                if choice==3:
                    for i in lst1:
                        if i<0:
                            print('Amount withdrawn : ',-1*i) # This will shows withdrawal amount history of current account
                        else:
                            print('Amount deposited : ',i)     # this will shows history of deposited amount of current account
                    print('Total balance : ',dict4[userid])        

                print('Do you want to do another transaction?') # user can start the transaction again from selecting account type
                choice=input('yes or no: ').lower()
                if choice=='yes':
                    continue                  # this loop will continue from 'select your account type'
                elif choice =='no':
                    print('Thank you for banking with us')
                    print('Rate our service from 1 to 5 :')
                    rate=int(input())
                    if rate<3:                 #giving option to user for suggestions if user rating is less than 3 out of 5
                        print('sorry for inconvinence :( , would you like to add any suggestion')
                        sugg=input('Add suggestion...')
                        print('Thank you')
                    elif rate>3:
                        print('Thank you! :D')
                    break                    # will break the while loop where user gets prompt for select account type 
        break                                # will break password while loop
    break                                    # will break the main outer while loop
                    

#Note: please refer dict2 and dict1 for userid and corresponding password(Ex:-userid:@grawal password:A01X9)

#dict2={'@grawal':'A01X9','roy@123':'RB25S','rav_321':'RU68R','anu&78':'AN34C','farh*345':'FN26D'}
#dict1={'A01X9':'Likhit','RB25S':'Bhushan','RU68R':'Ranveer','AN34C':'Anuj','FN26D':'Farahan'}
    
    
    
```
